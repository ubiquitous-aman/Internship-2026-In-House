## Dictionary

```python
marks = {
    "math": 90,
    "physics": 85,
    "chemistry": 88
}

print(f"Length of Dictionary is {len(marks)}")
print(f"Keys in the dictionary are: {marks.keys()}")
print(f"Values in the dictionary are: {marks.values()}")
print(f"Key-Value pairs in the dictionary is: {marks.items()}")

marks["english"] = 75 # add new value
marks["physics"] = 95 # update existing value

if "biology" in marks:
    print("Found")
else:
    print("Not Found")

for subject, score in marks.items():
    print(subject, score)

removed = marks.pop("english")
print("Removed:", removed)

new_marks = marks.copy()
print(new_marks)

marks.clear()
print(marks)
```

## Tuples

```python
fruits = ("apple", "banana", "mango", "banana")

print(f"Length of tuple is {len(fruits)}")
print(f"First element is: {fruits[0]}")
print(f"Last element is: {fruits[-1]}")

print(f"Count of banana is: {fruits.count('banana')}")
print(f"Index of mango is: {fruits.index('mango')}")

student = ("Aman", 20, "CSE")
name, age, branch = student
print(name)
print(age)
print(branch)

a = 10
b = 20
a, b = b, a # Swapping Operation
print(f"After swapping: a = {a}, b = {b}")

numbers = [1, 2, 3]
new_tuple = tuple(numbers)
print(new_tuple)

nested_tuple = (1, 2, (3, 4))
print(f"Nested tuple is: {nested_tuple}")
print(f"Element inside nested tuple is: {nested_tuple[2][1]}")
```

## Sets

```python
languages = {"Python", "C++", "Java"}

print(f"Length of set is {len(languages)}")
print(f"Initial set is: {languages}")

languages.add("Rust")
print(f"After adding Rust: {languages}")

languages.update(["Flutter", "JavaScript"])
print(f"After updating set: {languages}")

languages.remove("Java")
print(f"After removing Java: {languages}")

languages.discard("Swift")
print(f"After discarding Swift: {languages}")

if "Python" in languages:
    print("Python Found")
else:
    print("Python Not Found")

for language in languages:
    print(language)

a = {1, 2, 3}
b = {3, 4, 5}

print(f"Union: {a.union(b)}")
print(f"Intersection: {a.intersection(b)}")
print(f"Difference: {a.difference(b)}")
print(f"Symmetric Difference: {a.symmetric_difference(b)}")

nums = [1, 1, 2, 3, 3, 4]
unique_nums = set(nums)
print(f"Unique values are: {unique_nums}")

languages.clear()
print(f"Set after clearing: {languages}")
```

### 2) Write a function for basic math operations like add, multiply, subtract, divide and use this in your program, take 2 number input from user.

```python
def mathsOp(a, b):
    sum = a+b
    diff = a-b
    multiply = a*b
    if(b!=0):
        divide = a/b
    
    print(f"Sum of two numbers is {sum}.")
    print(f"Difference of two numbers is {diff}.")
    print(f"Product of two numbers is {multiply}.")
    print(f"Division of two numbers is {divide:.2f}.")

a = int(input("Enter 1st number: "))
b = int(input("Enter 2nd number: "))
mathsOp(a,b)
```

### 3) Write a program to check Palindrome Number.

```python
def checkPalindrome(n):
    if (n[0:len(n):1] == n[::-1]):
        print("This is a Palindrome.")
    else:
        print("Not a palindrome.")
    
n = (input("Enter a number: "))
print(f"User input recorded as {n}")
checkPalindrome(n)
```
