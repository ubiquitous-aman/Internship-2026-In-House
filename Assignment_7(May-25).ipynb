#### Different ways to iterate over rows in Pandas Dataframe 
#### Selecting rows in pandas DataFrame based on conditions 
#### Select any row from a Dataframe using iloc[]
#### Limited rows selection with given column
#### Drop rows from the dataframe based on certain condition applied on a column
#### Insert row at given position in Pandas Dataframe 
#### Create a list from rows in Pandas dataframe 

```python
# iterating through dataframe
bank_data = {
    'User':['Rohan', 'Rohit', 'Ranjeet', 'Raman'],
    'Branch':['Alwar', 'Jaipur', 'Jodhpur', 'Bikaner'],
    'Balance':[34500, 12000, 20000, 45000]
    }
df6=pd.DataFrame(bank_data)
```

```python
# Way 1
for cols in df6.columns:
    print(cols, df6[cols])
    print("_____________________________")
```

```python
# Way 2
for col in df6.items():
    print(col)
    print("_____________________________")
```

```python
# Way 3
for i in range(len(df6)):
    print(df6.iloc[i]['User'], df6.iloc[i]['Balance'])
print("______________________________________")
# Way 4
for i in df6.index:
    print(df6.loc[i, 'User'])
```

```python
#Selecting Rows based on condition
df6['Salaried'] = df6['Balance'] > 20000
```

```python
# Select 1st row (index 0)
print(df6.iloc[0])
print("______________________________________")
# Select 3rd row (index 2)
print(df6.iloc[2])
print("______________________________________")
# Select rows at index 0, 2, and 3
print(df6.iloc[[0, 2, 3]])
print("______________________________________")
# Select first 3 rows
print(df6.iloc[0:3])
print("______________________________________")
# Select rows from index 1 to end
print(df6.iloc[1:])
print("______________________________________")
# Select last 2 rows
print(df6.iloc[-2:])
print("______________________________________")
# Rows 0 to 2, and only 'User' and 'Balance' columns
print(df6.iloc[0:3, [0, 2]])
print("______________________________________")
# Rows 1 and 3, all columns
print(df6.iloc[[1, 3], :])
print("______________________________________")
# All rows, only Branch and Balance
print(df6.iloc[:, [1, 2]])
```

```python
# Drop rows where Balance < 25000
d = df6[df6['Balance'] > 35000].index
df6_n = df6.drop(d)
print(df6_n)
print("__________________________________________")
# Keep only rows where Balance >= 25000
df6_ne = df6[df6['Balance'] >= 25000]
print(df6_ne)
print("__________________________________________")
# Drop rows where Branch is 'Jaipur' OR Balance < 20000
condition = (df6['Branch'] == 'Jaipur') | (df6['Balance'] < 20000)
df_new = df6[~condition]
print(df_new)
```
