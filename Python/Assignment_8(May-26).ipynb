#### 2) Create two DataFrames, df1 and df2, with a common column (e.g., 'ID').
- Perform an inner merge on this common column and display the resulting DataFrame.
- Perform a left join of df1 and df2 on the 'ID' column. Demostrate how missing values are handled in the resulting DataFrame.
- Right Join and Index-Based Join
- Perform a right join using pd.merge() on a common column, then perform a join using df.join() based on the index. Compare the results. 
- Merging with Multiple Keys.

```python
students = {
        'stu_id': [4,5,1,3,2],
        'name':['Aman', 'Hitesh','Parvez', 'Anmol', 'Adarsh'],
        'age' : [18, 20, 25, 17, 22]
}
books = {
    'stu_id' : [4, 3,2,5],
    'Book_name' : ['The Hobbit', 'The Martian', 'Gone Girl', 'The Kite Runner'],
    'date_issued':['2026-01-23', '2026-02-10', '2026-02-10', '2026-01-15']
}

df0=pd.DataFrame(students, index=['S1', 'S2', 'S3', 'S4', 'S5'])
df1=pd.DataFrame(books, index=['B1', 'B2', 'B3', 'B4'])

df2 = df0.merge(df1, on='stu_id',how='inner')
print(df2)
print("____________________________________________________")
df3 = df0.merge(df1, on='stu_id', how='left')
print(df3)
print("____________________________________________________")
# Clearly, the missing values are filled with NaN
```

```python
df4 = pd.merge(df0, df1, on='stu_id', how='right')
print(df4)
print("____________________________________________________")
df5 = df0.join(df1, how='left', lsuffix='_stu', rsuffix='_book')
print(df5)
print("____________________________________________________")
df5 = df0.join(df1, how='right', lsuffix='_stu', rsuffix='_book')
print(df5)
print("____________________________________________________")
```

```python
idx1=df0.set_index('stu_id')
idx2=df1.set_index('stu_id')
df6=idx1.join(idx2, how='right')
print(df6)
print("____________________________________________________")
df6 = df0.join(df1, lsuffix='__s', rsuffix='__b')
print(df6)
```

Merging with multiple keys

```python
students2 = pd.DataFrame({
    'stu_id':[1,2,3],
    'semester':[5,5,6],
    'name':['Parvez','Adarsh','Anmol']
})

books2 = pd.DataFrame({
    'stu_id':[1,2,3],
    'semester':[5,5,6],
    'Book_name':['Python','DBMS','OS']
})

df7 = pd.merge(students2, books2, on=['stu_id', 'semester'],how='inner')
print(df7)
```

#### 3) Create three DataFrames. Vertically concatenate two of them using pd.concat(), then merge the resulting DataFrame with the third DataFrame on a common key to understand join() & merge().

- Also Explain the primary differences between df.join() and pd.merge(),

```python
students = pd.DataFrame({
    'stu_id': [1, 2, 3, 4, 5],
    'name': ['Aman', 'Hitesh', 'Parvez', 'Anmol', 'Adarsh'],
    'age': [18, 20, 25, 17, 22]
})
books = pd.DataFrame({
    'stu_id': [4, 3, 2, 5],
    'book_name': ['The Hobbit', 'The Martian', 'Gone Girl', 'The Kite Runner'],
    'date_issued': ['2026-01-23', '2026-02-10', '2026-02-10', '2026-01-15']
})
dept = pd.DataFrame({
    'stu_id': [1, 2, 3, 4, 5],
    'department': ['CSE', 'ECE', 'ME', 'CSE', 'EE'],
    'semester': [5, 5, 6, 4, 5]
})
```

```python
vertical = pd.concat(
    [students, books, dept],
    axis=0,
    ignore_index=True
)
print(vertical)
# When concatenating vertically, pandas creates the union of all columns. 
# For rows coming from one DataFrame, columns that don't exist in that DataFrame 
# are filled with NaN.
print("_______________________________________________________________")
new_merge = pd.merge(vertical, dept)
print(new_merge)
```

- `join()` : matches by index
- `merge()` : mathes by column name
