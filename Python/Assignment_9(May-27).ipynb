```python
df = pd.DataFrame(
    {
        'col1':['123_dgf', '%^4453'],
        'col2':('%@#^3527', '52@#7hd')
    }
)
print("Before: ")
print(df)
df['col1']=df['col1'].str.replace(r'\D+', '', regex=True).astype('int')
df['col2']=df['col2'].str.replace(r'\D+', '', regex=True).astype('int')
print("After: ")
print(df)
```

- A regex is all in one way to filter dataframes using 
    | Pattern  | Meaning            |
    | -------- | ------------------ |
    | \d       | digit              |
    | \D       | non-digit          |
    | \w       | word character     |
    | \W       | non-word character |
    | \s       | whitespace         |
    | ^abc     | starts with abc    |
    | xyz$     | ends with xyz      |
    | .        | any character      |
    | *        | 0 or more          |
    | +        | 1 or more          |
    | ?        | optional           |
    | [a-z]    | lowercase letters  |
    | [^0-9]   | not digits         |

```python
df = pd.DataFrame({
    "email": ["a@gmail.com", "b@yahoo.com", "c@gmail.com"]
})

mail = df[df["email"].str.contains(r"@gmail.com")]
print(mail)
print("________________________________")
s = pd.Series(["apple", "banana", "grape"])
print(s.str.contains(r"a"))   # containing one a
print(s.str.contains(r"a.a")) # exactly one char in between two a's
```

```python
df1 = pd.DataFrame({
    "date": ["2025-01-10", "2025-02-15", "2026-05-23", '2026-05-26']
})

df1["date"] = pd.to_datetime(df1["date"])
print(df1)
print(df1.dtypes)
print("_______________________")
df1["year"] = df1["date"].dt.year
print(df1)
df1["month"] = df1["date"].dt.month
print(df1)
df1["day"] = df1["date"].dt.day
print(df1)
df1["weekday"] = df1["date"].dt.day_name()
print(df1)
```
