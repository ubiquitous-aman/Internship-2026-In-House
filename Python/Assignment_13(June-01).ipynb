```python
a2 = np.array([[1, np.nan, 3],
                [4, 5, np.nan],
                [7, 8, 9]])

col_mean = np.nanmean(a2, axis=0)
inds = np.where(np.isnan(a2))
a2[inds] = np.take(col_mean, inds[1])
print(a2)
print("_______________________")
a3 = np.array([10, -5, 8, -2, 6])
a3[a3 < 0] = 0
print(a3)
print("_______________________")
a4 = np.array([3, 4])
a5 = np.array([1, 0])
avg = (a4 + a5) / 2
print("Average of NumPy arrays:")
print(avg)
print("_______________________")
a6 = np.array([[1, 2, 3],
                 [4, 5, 6]])
a7 = np.array([[7, 8, 9],
                 [10, 11, 12]])

a6_a7_combine = np.concatenate((a6, a7))

mean = np.mean(a6_a7_combine)
median = np.median(a6_a7_combine)
values, counts = np.unique(a6_a7_combine, return_counts=True)
mode = values[np.argmax(counts)]

print("Combined Array: ")
print(a6_a7_combine)

print("Mean =", mean)
print("Median =", median)
print("Mode =", mode)
```

```python
#  x - 2y + 3z = 9 
# -x + 3y -  z = -6 
# 2x - 5y + 5z = 17 

A = np.array([
    [1, -2, 3],
    [-1, 3, -1],
    [2, -5, 5]
])
B = np.array([9, -6, 17])
X = np.linalg.solve(A, B)

print("Solution:")
print("x =", X[0])
print("y =", X[1])
print("z =", X[2])

A_inv = np.linalg.inv(A)
X = np.dot(A_inv, B)
print(f"Solution using inverse of matrix: {X}")
```
