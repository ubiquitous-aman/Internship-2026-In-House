#### 1) Create a CSV file for address book, CSV file should have column for name, address, mobile, email. Insert 2-3 dummy data entered by user.


```python
import csv

n = int(input("Enter number of elements in list: "))
data_1 = [["Name", "Email", "Address", "Mobile"]]

for i in range(n):
    name = input("Name: ")
    email = input("Email: ")
    address = input("Address: ")
    mobile = input("Mobile: ")

    data_1.append([name, email, address, mobile])

with open("address_book.csv", "w", newline="") as file:
    writer = csv.writer(file)           #

    for row in data_1:
        writer.writerow(row)
```

### Practicing Database

##### Create a database

```python
import sqlite3

conn = sqlite3.connect("student.db")
```

##### Create tables

```python
sql1 = """
        CREATE TABLE student(
        stu_id INTEGER PRIMARY KEY,
        name VARCHAR(55) NOT NULL, 
        class VARCHAR(20),
        age INT NOT NULL)
        """
sql2 = """
        CREATE TABLE attendance (
        att_id INTEGER,
        stu_id INTEGER NOT NULL,
        date DATE NOT NULL,
        status TEXT NOT NULL)
        """

sql3 = """
        CREATE TABLE fees (
        fee_id INTEGER ,
        stu_id INTEGER NOT NULL,
        amount REAL NOT NULL,
        paid_on DATE)
        """
sql4 = """
        CREATE TABLE marks(
        english FLOAT NOT NULL,
        hindi FLOAT NOT NULL,
        maths FLOAT NOT NULL,
        science FLOAT NOT NULL)
        """
   
conn.execute(sql1)
conn.execute(sql2)
conn.execute(sql3)
conn.execute(sql4)
```

##### Perform Select operations

```python
s1 = conn.execute("SELECT * FROM student")
for row in s1:
    print(row)

s2 = conn.execute("SELECT * FROM attendance")
for row in s2:
    print(row)

s3 = conn.execute("SELECT * FROM marks")
for row in s3:
    print(row)
############################################
s4 = conn.execute(
    "SELECT * FROM student WHERE class='10A'")
for row in s4:
    print(row)
############################################
s5 = conn.execute("SELECT * FROM attendance")
for row in s5:
    print(row)
```

##### Perform Update on records

```python
conn.execute("""
    UPDATE student
    SET age = 18
    WHERE stu_id = 1
""")

conn.execute("""
    UPDATE fees
    SET amount = 4000
    WHERE stu_id = 3
""")

conn.commit()
```

##### Delete records 

```python
conn.execute("""
    DELETE FROM attendance
    WHERE att_id = 45
""")

conn.execute("DELETE FROM marks")
conn.commit()
conn.close()
```
