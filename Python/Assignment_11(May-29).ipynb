```python
a1 = np.array([23,45, 79, 64, 20])
print(a1)
print(f"Maximum in array is: {a1.max()}")
print("_______________________________")
a2 = np.array([23,45, 79, 64, 20])
print(a2)
print(f"Minimum in array is: {a1.min()}")
print("_______________________________")
a3 = np.array([[3,4,5,6,7],[2,4,6,8,0]])
print(a3)
print(f"No. of rows in array is: {(a3.shape)[0]}")
print(f"No. of columns in array is: {(a3.shape)[1]}")
print("_______________________________")
a4 = np.array([[31,42,53,64,75], [20,42,64,86,50], [51,63,85,47,79]])
print(f"Individual Array element (at 1st row) is: {a4[0]}")
print(f"Specific Array element (in 2nd row and 5th column) is: {a4[1][4]}")
print("_______________________________")
a5 = np.array([[31,22,63,94,25], [27,45,62,81,54], [51,63,85,47,79]])
print(f"Sum of all array elements(without for loop): {a5.sum()}")
row = (a5.shape)[0]
col = (a5.shape)[1]
sum = 0
for a in range(row):
    for b in range(col):
        sum += a5[a][b]
print(f"Sum of all array elements (using for loop): {sum}")
print("_______________________________")
a6 = np.array(range(10, 20, 2))
a7 = np.array(range(20, 40, 4))
print(f"Array 1: {a6}")
print(f"Array 2: {a7}")
print(f"Sum: {a6+a7}")
print(f"Difference: {a6-a7}")
print(f"Product: {a6*a7}")
a7_ = np.array(range(10, 30, 4))
print(f"Division: {(a6/a7_).round(2)}")
```

5. Iterate through a 3D NumPy array using:
-       Nested for loops
-       numpy.nditer()

```python
a8 = np.array([[[23,45,67],[32,54,76]],
               [[89,65,34],[28,91,18]],
               [[44,75,90],[34,57,86]]])

print(f"Printing 3D array (using for loop)")

for row in range((a8.shape)[0]):
    for col in range((a8.shape)[1]):
        for depth in range((a8.shape)[2]):
            print(f"Element{[row]}{[col]}{[depth]}: {a8[row][col][depth]}")
print("________________________________________")

print("________________________________________")
print(f"Printing 3D array (using nditer function)")

count = 0
for x in np.nditer(a8):
    count+=1
    print(f"Element[{count}]: {x}")
```

6. Study the following NumPy program and extend it to calculate statistical measures for two 2D arrays:
-       Mean (Average)
-       Median
-       Mode

```python
a9 = np.array([[31,22,44,94,25], [27,99,62,13,54], [51,63,15,47,18]])
a10 = np.array([[30,92,38,94,25], [27,45,37,80,36], [60,63,64,47,80]])

print(f"Array 1: {a9}")
print("___________________________")
print(f"Array 1: {a10}")
print("___________________________")
a11 = np.concatenate((a9.flatten(), a10.flatten()))
print(f"Combined Array: {a11}")
a12 = np.sort(a11)
print(f"Combined & Sorted Array: {a12}")
mean = np.mean(a11)
median = np.median(a11)
values, counts = np.unique(a11, return_counts=True)
mode = values[np.argmax(counts)]

print("Mean:", mean.round(2))
print("Median:", median)
print("Mode:", mode)
```
