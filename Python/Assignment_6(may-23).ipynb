### Open Weather API

```python
import requests


def get_weather(city):
    url = f"https://api.openweathermap.org/data/2.5/weather?q={city}&appid=d2a0678829009779a1aa0a465f2f5460&units=metric"
    try:
        response = requests.get(url)
        response.raise_for_status()
        data = response.json()
        print(data["main"]["temp"])
    except response.exceptions.RequestException as e:
        print(e)

city = input("Enter city: ")
s = get_weather(city)
```

### Rock-Paper-Scissor

```python
import random

try:
    consent = input("Enter yes/no: ")
except Exception:
    print("Enter a valid reply.")

u_score = 0
c_score = 0
if consent == "yes":
    print("User is ready to play.")

    n = int(input("Enter no. of rounds: "))
    while n:
        l = ["rock", "paper", "scissor"]
        print(f"The game is {n} rounds.")
        uch = input("Enter user choice: ")
        cch = random.choice(l)

        if (
            (uch == l[0] and cch == l[0])
            or (uch == l[1] and cch == l[1])
            or (uch == l[2] and cch == l[2])
        ):
            print(f"User & Computer chose {uch}")
            print("No points")
            n -= 1
        elif (
            (uch == l[0] and cch == l[1])
            or (uch == l[0] and cch == l[2])
            or (uch == l[2] and cch == l[1])
        ):
            print(f"User chose {uch}.")
            print(f"Computer chose {cch}.")
            print("User got the point.")
            u_score += 1
            n -= 1
        elif (
            (uch == l[1] and cch == l[0])
            or (uch == l[2] and cch == l[0])
            or (cch == l[2] and uch == l[1])
        ):
            print(f"Computer chose {cch}.")
            print(f"User chose {uch}.")
            print("Computer got the point.")
            c_score += 1
            n -= 1
    if u_score > c_score:
        print(f"User won the game. He scored {u_score} points.")
    elif u_score < c_score:
        print(f"Computer won the game. It scored {c_score} points.")
    else:
        print(f"Game draw. Both parties scored {c_score}.")

elif consent == "no":
    print("Game Quit")
```
