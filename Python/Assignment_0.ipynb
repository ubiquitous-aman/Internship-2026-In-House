1) Practice print statement: - with multiple object - with seperator - with "ends"

```python
name = "Aman"
age = 20
print("Name:", name, "Age:", age)
print("-------------------------")
print("Python", "Java", "C++", sep=" | ")
print(10, 20, 30, sep="-")
print("-------------------------")
print("Hello", end=" ")
print("World")
print("-------------------------")
print("Python", end="->")
print("Programming")
```
