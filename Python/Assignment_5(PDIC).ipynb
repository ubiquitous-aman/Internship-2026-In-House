1) Write a Python program to repeat a tuple three times using the * operator.

```python
t = (1, 2, 3)
result = t * 3

print("Repeated tuple:", result)
```

2) Write a Python program to join three separate tuples into one new tuple using the + operator.

```python
t1 = (1, 2)
t2 = (3, 4)
t3 = (5, 6)
new_tuple = t1 + t2 + t3

print("Joined tuple:", new_tuple)
```

3) Write a Python program to check whether a specific element exists inside a tuple using the in keyword.

```python
t = (10, 20, 30, 40)
element = 30

if element in t:
    print("Element found")
else:
    print("Element not found")
```

4) Write a Python program to calculate the total, highest value, and lowest value from a tuple of integers without using the built-in sum(), max(), and min() functions.

```python
numbers = (12, 45, 7, 23, 89, 34)

total = 0
highest = numbers[0]
lowest = numbers[0]

for num in numbers:
    total += num

    if num > highest:
        highest = num

    if num < lowest:
        lowest = num

print("Total =", total)
print("Highest =", highest)
print("Lowest =", lowest)
```

 5) Write a Python program to filter a tuple . n = (3, 14, 7, 22, 9, 41, 18, 5), keep only values greater than 10 

```python
n = (3, 14, 7, 22, 9, 41, 18, 5)
filtered = ()

for value in n:
    if value > 10:
        filtered += (value,)

print("Filtered tuple:", filtered)
```

6) Write a Python program to determine how many elements are in a set without using the built-in len() function. s = {"cat", "dog", "bird", "fish"}

```python
s = {"cat", "dog", "bird", "fish"}

count = 0

for item in s:
    count += 1

print("Number of elements:", count)
```

7) Write a Python program to combine two sets into one, containing all unique elements from both sets. 

```python
s1 = {1, 2, 3}
s2 = {3, 4, 5}
combined = s1 | s2

print("Combined set:", combined)
```

8) Write a Python program to find all elements that are common to both sets. s1 = {1, 2, 3, 4} s2 = {3, 4, 5, 6} 

```python
s1 = {1, 2, 3, 4}
s2 = {3, 4, 5, 6}
common = s1 & s2

print("Common elements:", common)
```

9) Write a Python program to find all elements that are in either set but not both. s1 = {1, 2, 3, 4} s2 = {3, 4, 5, 6} 

```python
s1 = {1, 2, 3, 4}
s2 = {3, 4, 5, 6}
result = s1 ^ s2

print("Elements in either set but not both:", result)
```
