### 1. Write a Python function to find the maximum of three numbers.

```python
def findMax(n1, n2, n3):
    if (n1 > n2) and (n1 > n3):
        print(f"{n1} is maximum number.")
    elif(n2 > n1) and (n2 > n3):
        print(f"{n2} is maximum number.")
    else:
        print(f"{n3} is maximum number.")

l=[]
for i in range(3):
    l.append(int(input("Enter numbers:")))
print(f"User input are {l[0]}, {l[1]}, and {l[2]}.")
findMax(l[0], l[1], l[2])
```

### 2. Write a Python function that takes a list and returns a new list with distinct elements from the first list.

```python
def distinctList(l):
    return list(set(l))  
   
l=[]        #defining an empty list
n = int(input("Enter length of List: "))
for i in range(n):
    l.append(int(input("Enter elements: ")))
distinctList(l)
```

### 3. Write a Python function to multiply all the numbers in a list.

```python
def multiplyNumber(l):
    m = 1
    for i in range(len(l)):
        m = m * l[i]
    print(m)

l=[]        #defining an empty list
n = int(input("Enter length of List"))
for i in range(n):
    l.append(int(input("Enter:")))
    print(f"User input are {l[i]}")
print(f"Multiplication is {multiplyNumber(l)}")
```

### 4. Write a Python function to calculate the factorial of a number (a non-negative integer). The function accepts the number as an argument.

```python
def factorial(n):
    if n==1: return 1
    return (n * factorial(n-1))

n = int(input("Enter a non-negative number"))
print(f"Factorial of {n} is {factorial(n)}.")
```

### 5. Write a Python program to reverse a string.

```python
def reverseString(s):
    return s[::-1]

s = input("Enter a string: ")
print(f"Reverse of {s} is {reverseString(s)}.")
```

### 7. Write a Python function to Print Even Numbers from a Given List

```python
def findEven(l):
    l1 = []
    for x in l:
        if (x%2 == 0): l1.append(x)
    return l1

l=[]        #defining an empty list
n = int(input("Enter length of List"))
for i in range(n):
    l.append(int(input("Enter elements in the list:")))
findEven(l)
```

### 8. Write a Python function that takes a number as a parameter and checks whether the number is prime or not.

```python
def checkPrime(n):
    if n <= 1:
        return False
    for i in range(2, n):
        if n % i == 0: return False
    return True

n = int(input("Enter a number: "))
if checkPrime(n):
    print(f"{n} is Prime Number")
else:
    print(f"{n} is Not a Prime Number")
```

### 9. Write a Python function that accepts a string and counts the number of upper and lower case letters. 

```python
def countUpperLowerChar(s):
    upper = 0
    lower = 0
    for x in s:
        if x.isupper(): upper += 1
        elif x.islower(): lower += 1
    print("Uppercase letters:", upper)
    print("Lowercase letters:", lower)

text = input("Enter a string: ")
countUpperLowerChar(text)
```
