```python
# 1) Write a python program that takes in a student name, class. It should also take in five subject marks of the students and 
#    find the total mark and percentage. Display a result in such a way that their name, class, and percentage are printed.

s_name = input("Enter the name of student: ")
s_class = input("Enter the student's class: ")

sub1 = int(input("Enter marks of 1st subject: "))
sub2 = int(input("Enter marks of 2nd subject: "))
sub3 = int(input("Enter marks of 3rd subject: "))
sub4 = int(input("Enter marks of 4th subject: "))
sub5 = int(input("Enter marks of 5th subject: "))

marks_sum = sub1 + sub2 + sub3 + sub4 + sub5
percentage = (marks_sum)/5

print(f"Name of the student is {s_name}.", end="")
print(f" His standard is {s_class}.")
print(f"Percentage obtained in five subjects is {percentage:.2f}.")
```

```python
# 2) Input two strings. Perform operations such as concatenation and store it in another variable. Demonstrate functions such as lower(), upper(), 
#    title(), swapcase(), capitalize(), casefold(), center(), count(), endswith(), find(), isalnum(), isdigit(), isnumeric(), isspace(), and replace()

str1 = input("Enter first string: ")
str2 = input("Enter second string: ")

str_new = str1 + str2
print(f"Concatenated String is {str_new}.")

print("Using lower():", str_new.lower())
print("Using upper():", str_new.upper())
print("Using title():", str_new.title())
print("Using swapcase():", str_new.swapcase())
print("Using capitalize():", str_new.capitalize())
print("Using casefold():", str_new.casefold())
print("Using center():", str_new.center(40))
print("Count of 'm':", str_new.count('m'))
print("Ends with 'n':", str_new.endswith("n"))
print("Position of 'o':", str_new.find('o'))
print("Is alphanumeric:", str_new.isalnum())
print("Is digit:", str_new.isdigit())
print("Is numeric:", str_new.isnumeric())
print("Is space:", str_new.isspace())
print("After replace():", str_new.replace('Python', 'C++'))
```

```python
# 3) Demonstrate the use assignment operators in Python.

x = int(input("Enter a number: "))

print("Initial value:", x)

x += 5
print("After += 5:", x)

x -= 3
print("After -= 3:", x)

x *= 2
print("After *= 2:", x)

x /= 4
print("After /= 4:", x)

x %= 3
print("After %= 3:", x)

x **= 2
print("After **= 2:", x)

x //= 2
print("After //= 2:", x)
```
