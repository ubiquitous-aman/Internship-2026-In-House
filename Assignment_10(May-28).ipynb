1) Replace Nan with 0 and Interchange 3 rows and 3 columns of 2D array [[6, -8, 73, -110], [np.nan, -8, 0, 94]] 
2) Move axes of 3D array to new positions 
3) Replace NaN values with average of columns 
4) Replace negative value with zero in numpy array using replace

```python
import numpy as np

arr = np.array([[6, -8, 73, -110], [np.nan, -8, 0, 94]])
print(arr)
print("____________________________")
arr1 = np.nan_to_num(arr, nan=0.0)
print(arr1)


#Interchanging the columns
arr[[0, 1]] = arr[[1, 0]]
arr[:, [2, 3]] = arr[:, [3, 2]]
```

```python
arr2 = np.random.randint(0, 100, size=(2, 3, 4))
print(arr2.shape)

# change axes
arr4 = np.moveaxis(arr2, [0, 1, 2], [1, 2, 0])
#arr4 = np.transpose(arr2, (1,2,0))
print(arr4.shape)
```

```python
arr5 = np.array([[1.0, 2.0, np.nan],
                [4.0, np.nan, 6.0],
                [7.0, 8.0, 9.0]])
print(arr5)          

col_means = np.nanmean(arr5, axis=0)
for i in range(arr5.shape[1]):
    arr5[:, i] = np.where(np.isnan(arr5[:, i]), col_means[i], arr5[:, i])
print("_________________")
print(arr5)
```

```python
arr6 = np.array([[6, -8, 73, -110],
                [-5, -8, 0, 94],
                [2, -3, 10, -1]])
print(arr6)
arr6 = np.where(arr6 < 0, 0, arr6)
print("_________________________")
print(arr6)
```
